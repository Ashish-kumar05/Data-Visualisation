```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Inbuilt Data set in Seaborn = anscombe, attention, brain_networks, car_crashes, dots,
# exercise, flights, fmri, gammas, iris', planets, tips, titanic
```

```python
# Seaborn Line Plot

# sns.lineplot(x='', y='', data=)

Data = sns.load_dataset("fmri")

print(Data)
```

```python
A = sns.lineplot(x= "timepoint", y="signal", data = Data)
print(A)
```

```python
sns.lineplot(x= "timepoint", y="signal", data = Data)
plt.show()
```

```python
# Grouping Data with "hue" command

sns.lineplot(x= "timepoint", y="signal", data = Data, hue = "event")
plt.show()
```

```python
# Adding line style with "style" command

sns.lineplot(x= "timepoint", y="signal", data = Data, hue = "event", style = "event")
plt.show()
```

```python
# Adding Markers

sns.lineplot(x= "timepoint", y="signal", data = Data, hue = "event", style="event", markers= True)
plt.show()
```

```python
Data1 = sns.load_dataset("titanic")
Data1.head(15)
```

```python
# Bar plot

sns.barplot(x="who", y="pclass", data=Data1)
plt.show()
```

```python
sns.barplot(x="who", y="pclass", data=Data1, hue="class")
plt.show()
```

```python
# Want changing the color of both bars- using 'palette' command ("Blues_d", "vlag", "rocket")

sns.barplot(x="who", y="pclass", data=Data1, palette= "rocket")
plt.show()
```

```python
sns.barplot(x="who", y="pclass", data=Data1, color= "green") # All bars have same color
plt.show()
```

```python
MyData= sns.load_dataset("iris")
print(MyData)
```

```python
# Scatter Plot

sns.scatterplot(x="sepal_length", y="petal_length", data= MyData)
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data= MyData, hue="species")
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data= MyData, hue="species", style="species")
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data= MyData, hue="petal_length")
plt.show()
```

```python
# Histogram/ Distplot(Distribution)

# distplt() function

Mydata= sns.load_dataset("iris")

print(Mydata)

```

```python
sns.displot(Mydata["sepal_length"], kde= True, color="red") # [kde(for pogygon frequency) = True, for remove= false]
plt.grid()
plt.show()
```

```python
# [show graph on y axis]
sns.distplot(Mydata["sepal_length"], color="green", vertical=True) 

plt.show()
```

```python
# Joint Plot

sns.jointplot(x="sepal_length", y="petal_length", data=MyData, color ="Brown")
plt.show()
```

```python
# Joint Plot - Add regression Line

sns.jointplot(x="sepal_length", y="petal_length", data=MyData, color ="olive", kind= "reg")
plt.show()
```

```python
# Box Plot

MyData.head(2)
```

```python
sns.boxplot(x="species", y="sepal_length", data=MyData, palette="Set1")
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_length", data=MyData, palette="Set1", linewidth= 5)
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_length", data=MyData, order=["versicolor", "virginica", "setosa"])
plt.show()
```
